```python
def calculate_sparsity(df):
    """
    Calculate the percentage of sparsity in a DataFrame.

    Parameters:
    df (pandas.DataFrame): The DataFrame to analyze.

    Returns:
    float: The percentage of sparsity in the DataFrame.
    """
    # Calculate the total number of elements in the dataframe
    total_elements = df.size

    # Calculate the number of zero or NaN values in the dataframe
    zero_or_nan_elements = (df == 0).sum().sum() + df.isna().sum().sum()

    # Calculate the percentage of sparsity
    sparsity_percentage = (zero_or_nan_elements / total_elements) * 100

    return sparsity_percentage
```

```python
# Example usage
sparsity_percentage = calculate_sparsity(df)
print(f"Percentage of sparsity in the dataframe: {sparsity_percentage:.2f}%")
```

```python
# Directory containing the files
directory = '/gpfs/gibbs/pi/krishnaswamy_smita/jwk47/splice/data/expression/mi_raw/'

# List to store sparsity values
sparsity_values = []

# Iterate over all files in the directory
for file in os.listdir(directory):
    if file.endswith('.csv'):
        # Load the file into a DataFrame
        df = pd.read_csv(os.path.join(directory, file), index_col=0)
        
        # Apply the lambda function to create the adjacency matrix
        adj_mat = df.applymap(lambda x: 1 if float(x) > 0.35 else 0)
        
        # Calculate the sparsity percentage
        sparsity_percentage = calculate_sparsity(adj_mat)
        
        # Extract the unique file label
        file_label = file.split('_')[2]
        
        # Print the sparsity percentage with the file label
        print(f"Percentage of sparsity in the adjacency matrix for {file_label}: {sparsity_percentage:.2f}%")
        
        # Store the sparsity value in the list
        sparsity_values.append((file_label, sparsity_percentage))

sparsity_values
```

```python
import matplotlib.pyplot as plt

# Extract the sparsity percentages from the sparsity_values list
sparsity_percentages = [value[1] for value in sparsity_values]

# Create a histogram of the sparsity percentages
plt.hist(sparsity_percentages, bins=20, edgecolor='black')
plt.title('Histogram of Sparsity Percentages')
plt.xlabel('Sparsity Percentage')
plt.ylabel('Frequency')
plt.show()
```
