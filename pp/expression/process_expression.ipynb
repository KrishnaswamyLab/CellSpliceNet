```python
import pandas as pd

df_magic = pd.read_csv('../../data/expression/counts/240919_scCounts_neuronsBulk_magic_filtered.csv', index_col=0)
df_magic
```

```python
# Split each element in col_headers on '_' and take the first part
split_headers = [header.split('_')[0] for header in df_magic.columns]

# Find unique elements
unique_headers = set(split_headers)

# Print the unique elements
print(unique_headers)
print(f"Number of unique elements: {len(unique_headers)}")
```

```python
# Create a dictionary to store DataFrames for each unique header
header_dfs = {}

# Loop through each unique header
for header in unique_headers:
    # Filter columns that include the current header
    filtered_cols = [col for col in df_magic.columns if col.startswith(header)]
    
    # Create a new DataFrame with the filtered columns
    header_dfs[header] = df_magic[filtered_cols]

# Print the keys of the dictionary to verify
print(header_dfs.keys())
```

```python
# Create a new DataFrame to store the means
mean_df = pd.DataFrame()

# Loop through each key-value pair in header_dfs
for header, df in header_dfs.items():
    # Calculate the mean for each row and add it as a new column in mean_df
    mean_df[header] = df.mean(axis=1)

mean_df
```

```python
mean_df.to_csv('../../data/expression/counts/expression_average_raw.csv', index=True)
```

```python
from scipy.stats import skew, kurtosis

# Create a dictionary to store statistical moments for each DataFrame
moments_dfs = {}

# Loop through each key-value pair in header_dfs
for header, df in header_dfs.items():
    # Calculate mean, variance, skewness, and kurtosis for each row
    mean = df.mean(axis=1)
    variance = df.var(axis=1)
    skewness = df.apply(skew, axis=1)
    kurt = df.apply(kurtosis, axis=1)
    
    # Create a new DataFrame to store the moments
    moments_df = pd.DataFrame({
        'mean': mean,
        'variance': variance,
        'skewness': skewness,
        'kurtosis': kurt
    })
    
    # Store the moments DataFrame in the dictionary
    moments_dfs[header] = moments_df

# Print the keys of the dictionary to verify
print(moments_dfs.keys())
```

```python
import os

# Create the directory if it doesn't exist
output_dir = '../../data/expression/moments'
os.makedirs(output_dir, exist_ok=True)

# Loop through each key-value pair in moments_dfs and save to CSV
for key, moments_df in moments_dfs.items():
    filename = f"{key}_moments.csv"
    filepath = os.path.join(output_dir, filename)
    moments_df.to_csv(filepath, index=True)
```

```python
from sklearn.feature_selection import mutual_info_regression
from tqdm import tqdm


def build_mi_graph(df_subset):
    df_subset = df_subset.T

    # Initialize an empty DataFrame to store mutual information values
    mi_matrix = pd.DataFrame(index=df_subset.columns, columns=df_subset.columns)
    # Calculate mutual information for each pair of columns
    for col1 in tqdm(df_subset.columns, desc="Outer loop"):
        for col2 in tqdm(df_subset.columns, desc="Inner loop", leave=False):
            if col1 != col2:
                mi = mutual_info_regression(df_subset[[col1]], df_subset[col2])
                mi_matrix.loc[col1, col2] = mi[0]
            else:
                mi_matrix.loc[col1, col2] = 0  # Mutual information with itself is 0

    print(mi_matrix)
```

```python
# Loop through each key-value pair in header_dfs
for header, df_subset in header_dfs.items():
    # Perform operations on each DataFrame
    print(f"Header: {header}")
    build_mi_graph(df_subset)
    break
    
```

```python
import os
import pandas as pd

# Directory containing the files
directory = '../../data/expression/mi_raw'

# Initialize variables to store the global max and min values
global_max = float('-inf')
global_min = float('inf')

# Loop through each file in the directory
for filename in os.listdir(directory):
    if filename.endswith('.csv'):
        filepath = os.path.join(directory, filename)
        df = pd.read_csv(filepath, index_col=0)
        
        # Find the max and min values in the current file
        file_max = df.max().max()
        file_min = df.min().min()
        
        # Update the global max and min values
        if file_max > global_max:
            global_max = file_max
        if file_min < global_min:
            global_min = file_min

print(f"Global maximum value: {global_max}")
print(f"Global minimum value: {global_min}")
```
